```python

import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

# Load stop words
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters
    text = ' '.join([word for word in word_tokenize(text) if word not in stop_words])  # Remove stop words
    return text
```

```python
D1['text'] = D1['text'].apply(preprocess_text)
```

```python
D1
```

```python
D1 = D1.rename(columns={'airline_sentiment':'sentiment'})
```

```python
D1['sentiment'] = D1['sentiment'].map({'positive': 2, 'neutral': 1, 'negative': 0})
```
